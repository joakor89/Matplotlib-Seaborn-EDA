### There are other types of graphs that Matplotlib provides in order to perform more accurate analysis.

Let's import the regular packages & methods

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Let's set up some variables to work with:

```python
data = np.random.randint(1,50,100)
```

#### Let's make a Histogram

---
*   Bins: focus on set a determine number of bars
*   histtype: focus o set or change the bar style

```python
plt.hist(data,bins=10, histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10, histtype='step')
plt.show()
```

### Boxplot

---
*   Patch Artist: Change direction, fill inter quartile range
*   Notch: focus median in the Boxplot
*   Showfliers: Takes away the outliers
*   List item


```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, vert=False ,patch_artist=True, notch=True)
plt.show()
```

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False ,patch_artist=True, notch=True, showfliers=False)
plt.show()
```

# Scatter Plots

Let's now set up some variables and perform these type of graphs

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area, c=colors)
plt.show()
```

### Let's add markers and some transparency in order to chechout any overlapping data within:

```python
plt.scatter(x,y,s=area, c=colors, marker='H', alpha=0.5)
plt.show()
```
