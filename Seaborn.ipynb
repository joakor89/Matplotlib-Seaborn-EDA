### Seaborn Graph Types

---
Seaborn offers certain main features for specific data visualization problems:

1.   Relatable diagrams or graphs
1.   Data distribution
2.   Graph categorical variables

For instance:


*   Relplot (relational): scatterplot, lineplot.
*   Displot (distributions): histplot, kdeplot, ecdfplot, rugplot.
*   Catplot (categorica): stripplot, swamplot, boxplot, violinplot, pointplot, barplot.
