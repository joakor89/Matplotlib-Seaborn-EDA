# Distributions 

---
### Seaborn offers some graphs or options that allow to work with distributions oriented to numeric data types. Let's see how you can achieve it using code.

### Let's import the required packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Performing graph with .kde & .displot

---
1.   Cumulative: Focus on how gather(cumulate) the distributions values & its behaviour.
1.   Stats: It wont' take count as parameter, instead, it will focus on new's parameter set up. for instance 'frequency'. This could change like 'probability' and so on.
2.   Multiple: Offers to shown the constrast between any distribution but tends to stack the visual information. For isntance: dodge, layer, stacks, fill, etc.
2.   Kind: Focus on change the class type in which the graph will show the distribution information.
1.   Shade: It will overlap or contrast the distribution by broadcasting a shadow on it

```python
tips = sns.load_dataset("tips")
tips.head()
```

### Let's now check how to work with distribution graphs:

```python
sns.histplot(data=tips, x='tip', bins=15)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='frequency')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15,cumulative=False, hue='sex', stat='percent')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15,cumulative=False, hue='sex', stat='density',multiple='stack')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15,cumulative=False, hue='sex', stat='density',multiple='layer')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15,cumulative=False, hue='sex', stat='density',multiple='dodge')
plt.show()
```

### Using .kde:

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde', multiple='stack')
plt.show()
```
