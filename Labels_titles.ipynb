# Legends, Labels, titles & size

---

*   In order to give context to the graphs, it needs to be used Titles, Legends, Size or Labels, so that our graphs have a broader context to show and even explain itself.

### How to give context to charts
*   Create a common ground for graphics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'lightgreen')
ax2.plot(y,x, 'salmon')
plt.show()
```

### Charts context with titles

To create the titles do the following:

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'lightgreen')
ax1.set_title("Relation X - Y")

ax2.plot(y,x, 'salmon')
ax2.set_title("Relation Y - X")
fig.show()
```

### Adding Context Tags

---
Adding labels on the Axes


```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'salmon', )
ax1.set_title("Relation X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(y,x, 'yellow')
ax2.set_title("Relation Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
fig.show()
```

### Resize shapes
Take the following action:

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (5,5))
ax1.plot(x,y, 'salmon', )
ax1.set_title("Relation X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(y,x, 'yellow')
ax2.set_title("Relation Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
fig.show()
```

### Applying Labels 

---
| It set up the best option by default |

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y, 'green', label = "$sin(x)$")
ax1.set_title("Relation X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
fig.show()
```

```python
fig, axes = plt.subplots(1,2, figsize= (5,5))
axes[0].plot(x,y, 'salmon', label="$sin(x)$")
axes[0].set_title("Relation X - Y")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend(loc='lower left')

axes[1].plot(y,x, 'yellow', label="$sin(y)$")
axes[1].set_title("Relation Y - X")
axes[1].set_xlabel("Y")
axes[1].set_ylabel("X")
axes[1].legend(loc='upped right')
fig.show()
```

### Let's try to run it out with Matplot instead of Object-Oriented base

```python
plt.plot(x,y,'salmon', label='sin(x)')
plt.title('A Title')
plt.xlabel('X axe')
plt.ylabel('Y axe')
plt.legend(loc='lower left')
fig.show()
```

```python
plt.plot(x,y,'#0343DF', label='sin(x)')
plt.title('A Title')
plt.xlabel('X axe')
plt.ylabel('Y axe')
plt.legend(loc='lower left', bbox_to_anchor=(1,0.2))
fig.show()
```
