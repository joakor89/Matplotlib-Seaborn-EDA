# subplot

---
*   Subplot allows you to create plots within a plot.
*   It does this via graph's array and can be accessed via index:

### How to access Plot-Matrix in Subplot:

---
*   Let's import our modules & set up some variables

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

### Parameters at Subplot

---
The subplot parameters work: | Rows | - | Columns | - | Index |


```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g--')

plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

### I've decided to run out the previous example, nevertheless:
---

*   I've added a new subplot line to the first graph 
*   On the second one, I've run out a pie chart instead of a histogram

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g--')
plt.subplot(1,2,1)
plt.plot(y,x, 'b:')

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'y-')
plt.plot(y,x, 'b:')

plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
