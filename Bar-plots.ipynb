# Bar Plots

---
The bar chart allows to plot categorical variables, that is, text variables, which is very important within Data Science's eviroment and Matplotlib offers certain features that make data scientist's lives easier in terms of plotting these types of variables.

### Let's import the method packages and set up some list, if necessary:


```python
import numpy as np
import matplotlib.pyplot as plt
```

Categorical variables:

```python
countries = ["India", "Japan", "Mexico", "Colombia", "Germany"]
population = [1408, 125.7, 126.7, 51.52, 83.2]
```

### Let's set a Bar Graph:

```python
plt.bar(countries, population)
plt.show()
```

##### Lets modify Bar Color & Size:

```python
plt.bar(countries, population, width=0.5, color = ["#BBE90F","#EE82EE","#FFA500","#008080","#FC5A50"])
plt.show()
```

##### It finally set Title, x & y labels for this graph

```python
plt.bar(countries, population, width=0.5, color = ["#BBE90F","#EE82EE","#FFA500","#008080","#FC5A50"])
plt.xticks(np.arange(5), ("India", "Japan", "Mexico", "Colombia", "Germany"), rotation=45)
plt.title('Population by Country')
plt.xlabel('Countries')
plt.ylabel('Population')
plt.show()
```

##### On simpliest way, The same bar graph can be horizontally performed

```python
plt.barh(countries,population, color = ["#BBE90F","#EE82EE","#FFA500","#008080","#FC5A50"])
plt.show()
```
