# Heatmap

---
*   Heatmap is a graph type focused on Matrix Structure. 
*   Heatmap maps all the numeric variables in the dataset.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

### Variables correlation

```python
tips.corr()
```

Seaborn offers Heatmap

---
*   Annot: Shows the correlation values
*   Cmap: shows & allows to change the Heatmap color
*   List item
*   List item



```python
sns.heatmap(tips.corr())
plt.show()
```

### Default set up

```python
sns.heatmap(tips.corr(), annot=True, cmap='viridis_r', linewidths=5, linecolor='yellowgreen')
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='viridis_r', linewidths=5, linecolor='yellowgreen', vmin=0, vmax=1)
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='viridis_r', linewidths=5, linecolor='yellowgreen', vmin=-1, vmax=1)
plt.show()
```
