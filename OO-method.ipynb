# Object-Oriented Method in Matplotlib

---

*   An object defines a figure, that figure is a canvas in which we can introduce different graphs (axes), each of which has its own axes (axis).
*   The figure represents the whole, within it we are going to configure graphs which contain different axes.
*   It's a bit more complicated, but on the same graph we can customize it much better.

### Main differences between Pyplot and Object Oriented

    Pyplot     >      Object Oriented

    Fast              Greater customization
    Easy              More code
    A single figure   More friendly to multiple diagrams

### Lets call out & import the methods


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

### Axe Parameters:
---
    |Pos. x-axis | | Pos. y-axis | | Size graph x-axis | | Size graph y-axis |


*   Now, let's focus to perform on object concept and create a more customizable graph:

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'y')
fig.show
```

### We can create additional graphs within the canvas by calling the fig. method:

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.17,0.55,0.4,0.3])

axes.plot(x,y,'g')
axes2.plot(y,x,'r')
axes2.set_facecolor('pink')
plt.show()
```
