# Pyplot 

---

*   Pyplot is a tool Matplotlib has to run graphs in a simple way. Let's see how you can achieve it.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Let's set up the variables to perform the analysis with:

```python
x = np.linspace(0,5,11)
y = x ** 2
x
y
```

### Let's plotting on Y function

```python
plt.plot(x,y)
plt.show()
```

### It can change the color as desired

---
As you can look at the third parameter
*   The first letter assigned the color
*   The second one assigned the linesharp or way to plot out

```python
plt.plot(x,y,'yx')
plt.show()
```

### Same exercise. Just reset the color and way to show the plotting

```python
plt.plot(x,y,'gs-')
plt.show()
```

```python
plt.plot(x,y,'yD-')
plt.show()
```

### Modify graphs using Pyplot

*   Now, we are going to change the color of the graph

#### Character	>   color
      ‘b’	         blue
      ‘g’	         green
      ‘m’	         magent



### Change more parameters

*   Strings Format

### Character	   >         Description

    ‘.’	              point marker
    ‘,’	              pixel marker
    ‘o’	              circle marker
    ‘v’	              triangle_down marker
    ‘^’	              triangle_up marker
    ‘<’	              triangle_left marker
    ‘>’	              triangle_right marker




### Line Styles

### Character    >     Description

    ‘-’                solid line style
    ‘–’                dashed line style
    ‘-.’               dash-dot line style
    ':'                dotted line style

### Let's set up a Histogram

```python
plt.hist(x)
plt.show()
```

### Let's set up a Pie Chart

```python
plt.pie(x)
plt.show()
```

### How to recognize correlation between variables

```python
plt.scatter(x,y, color='orange')
plt.show()
```

### Data Distribution

```python
plt.boxplot(x)
plt.show()
```

```python
plt.boxplot(y)
plt.show()
```
