# Seaborn Set

---
Set is a Seaborn method that allows you to set the style, font, and chart colors.

Let's import the required packages:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### BarPlot Graph

```python
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show()
```

### Style & Font Settings

---
What do you need to know when using Set?

*   This parameter affects all charts created it, including ones that don't use Seaborn.
*   We can reset the values using the reset_orig method.
*   This parameter is a shorthand for set_theme, which is recommended by the documentation.

```python
sns.set(style='darkgrid', palette="viridis_r", font="Verdana", font_scale=1)
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show()
```
