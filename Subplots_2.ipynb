# Subplots

### You can work with subplots on graph arrays that are accessed through the index.
---

### How to work with subplots for plotting graphs:
*   Let's import the method to be use & set up some variables to work with



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

### Create a Fig-like graph

---
*   To create a type Fig. graph, then you need to work with the following code:






```python
fig, axes = plt.subplots()
axes.plot(x,y,'y')
plt.show()
```

### Let's perform multiple fig.type graphs:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'g')
axes[1].plot(y,x,'r')
plt.show()
```

### Access charts by name stored within a tuple:
I just previously defined the arrays exit

```python
fig, (axe1,axe2) = plt.subplots(nrows=1, ncols=2)
axe1.plot(x,y,'y')
axe2.plot(y,x,'b')
plt.show()
```

### We are going to generate a type fig's matrix, with two rows and four columns:

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2,4)
ax1.plot(x, np.sin(x), 'lightblue')
ax2.plot(y, np.cos(x), 'violet')
ax3.plot(x, np.tan(y), 'orange')
ax4.plot(y, np.cos(y)**2, 'salmon')

fig.tight_layout()
```

### Let's finish it...

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x),'g')
axes[0,2].plot(x, np.tan(x),'y')
axes[0,3].plot(x, np.cos(x)**2,'r')

axes[1,0].plot(x, np.tan(x)**2,'salmon', linestyle='-')
axes[1,1].plot(x, x**3, 'lightblue', linestyle='--')
axes[1,2].plot(x, y**2, 'brown')
axes[1,3].plot(x*2.5, y, "g:")
```
