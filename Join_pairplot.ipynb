# Joinplot & Pairplot

---
They're Seaborn functionalities which  with a single command you can generate several types of graphs.

Let's import the packages required & set up the dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

# 1. Jointplot

---
Mainly gather two different graph as one that allows to take away more valueble analytical information.

Consequently, Let's analyze a cross-combination between dispersion and distribution diagram charts.


```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

### Let's improve & customize them

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

Marginal_ticks shows a little extra scale for the histogram of the graph. This can make noise when the graph is observed, but this tool will depend of the context

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25, fill=False, multiple='dodge'))
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

### Pair plot

---
*   Focus to show the relationship among numerical variables.
*   Takes all the numerical value and contrast them all to look at how the relationship work among them.

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True)
plt.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.show()
```
