# Categorical variables on Seaborn

---


*   Seaborn offers different graph types to graph categorical data or text variables. 
*   Below you will see the code that you can use to graph this type of data.

### Categorical data types to graph in Seaborn
Start by importing the data you need from the libraries:


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

### Quantity Graph

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

It can also by performed horizontally by just changing the axe to look up

```python
sns.countplot(data=tips, y='day', hue='sex')
plt.show()
```

This graph can be seen complex on first sight, then, let's add more context in order to make it more digestable:


*   "dodge": Allows to perperdicularly split the data in order to make easier to analyze it 
*   List item



```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

### Let's now check the swarmplot:

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

### Let's check out the BoxPlot
---
This will show the statistical distribution


```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

# It can be focus on numerical rather than categorical values

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, y='total_bill')
plt.show()
```

### Or even combine features in order to get cross-undestanding

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color="0", marker='<')
plt.show()
```

### It can also perform a Violin Plot:

```python
plt.figure(figsize=(8,8))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='bright')
plt.show()
```

Let's check a numerical variable with Violin Plot:

```python
sns.violinplot(data=tips, y='total_bill')
plt.show()
```

### Finally, let's move towards CatPlot:

---
This will perform similarly as stripplot broacast & with categorical variables.


```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, kind='box',col='time')
plt.show()
```
