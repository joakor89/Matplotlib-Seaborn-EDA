# Color & Styles 

### You can better customize the graphics with different colors and styles, thus, our graphics will be much better understood.
---

```python
import numpy as np
import matplotlib.pyplot as plt 
```

### Let's set up a quick variable to perform the visualization

```python
x = np.linspace(0,5,11)
```

# Now! lets check which Graph style we do have available:
This ought to show the different pre-define style set that we can work on

```python
print(plt.style.available)
```

### Setting Grip type for the Viz..

```python
plt.style.use('seaborn')
```

### Matlab modifications:

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1,'salmon', marker='o')
ax.plot(x, x+2,'y--')
ax.plot(x, x+3,'b.-')
ax.plot(x, x+4,'g:')
fig.show()
```

### Pyplot Performance:

*   The color can be assigned by color palette name or
*   It can also be assigned by RGB serial number
*   Alpha parameter will set up a certain level of line transparency. Lower the level, the more transparency will show up
*   Linewidth will focus on line-thinkness

*   Marker
*   Markersize 
*   Markerfacecolor
*   List item





```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, color='yellowgreen', alpha= 0.5, linewidth=18)
ax.plot(x, x+2, color='goldenrod', linewidth=3)
ax.plot(x, x+3, color='salmon', linewidth=5)
ax.plot(x, x+4, color='#0343DF', alpha= 0.3, linewidth=12)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, color='yellowgreen', alpha= 0.9, linewidth=2, linestyle='--', marker='o', markersize=8, markerfacecolor='violet')
ax.plot(x, x+2, color='yellow', linewidth=2,linestyle='--', marker='o', markersize=8, markerfacecolor='violet')
ax.plot(x, x+3, color='salmon', linewidth=2,linestyle='dashed', marker='o', markersize=8, markerfacecolor='green')
ax.plot(x, x+4, color='#0343DF', alpha= 0.5, linewidth=2,linestyle='-', marker='P', markersize=8, markerfacecolor='orangered')
fig.show()
```
